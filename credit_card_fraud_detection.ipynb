```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
print("Shape: ",credit_card_data.shape)
```

```python
# Checking the no. of missing values
credit_card_data.isnull().sum()
```

```python
# replacing the NULL value with empty string
credit_card_data = credit_card_data.fillna(0)
credit_card_data.isnull().sum()
```

```python
# Distribution of legit transactions & fradulant transaction
credit_card_data['Class'].value_counts()
```

This Dataset is highly unbalanced

0 --> Normal Transaction
1 --> Fraud Transaction

```python
# Seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]

print(legit.shape)
print(fraud.shape)
```

```python
# Statistical Measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

**Under-Sampling**

Build a sample dataset containing distribution of normal transaction and fradulent transaction

No. of Fraud transaction - 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating 2 dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
print(Y)
```

Split the data into training Data & test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model training**

Logistic Regression

```python
model = LogisticRegression()
```

Training the Logistic Regression model with training data

```python
model.fit(X_train, Y_train)
```

**Model Evaluation**

Accuracy Score

```python
# accuracy on Training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print("Training Accuracy:", training_data_accuracy)
```

```python
# accuracy on Testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print("Testing Accuracy", testing_data_accuracy)
```
